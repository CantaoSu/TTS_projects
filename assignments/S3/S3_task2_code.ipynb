```python
with open('S3_task2_original', 'r') as file:
    lines = file.readlines()

# Get the EN phoneme list from eSpeak GitHub
en_phonemes = ('p', 'p_h', 't', 't_h', '4', 'k', 'k_h', 'b', 'd', 'g',
    'f', 's', 'S', 'tS', 'T', 'v', 'z', 'Z', 'dZ', 'D',
    'm', 'n', 'N', 'l', 'r', 'j', 'w', 'h', 'r=', 'i',
    'A', 'O', 'u', 'I', 'E', '{', 'V', 'U', '@', 'EI',
    'AI', 'OI', '@U', 'aU', '_', ' ', '{')

# Define a dictionary to map Dutch eSpeak transcription to English eSpeak transcription based on my observation into IPA
replace_rules = {
    'I:': 'e',
    'O:': 'O',
    'Y': 'i',
    'Au': 'aU',
    'a': 'A',
    'x': 'h',
    'G': 'h',
    'nj': 'n',
    'W': 'w',
    'dj': 'dZ',
    '2': 'E',
    'y': 'i',
    '9y': 'aU',
    'tj': 'tS',
    'Y~': '@',
    'L': 'l',
    'J': 'j',
    'R': 'r=',
    'A~': 'A',
    'Oe': '@U',
    'e': 'E',
    'E~': 'E',
    'o': 'O',
    'O~': 'O',
    'Y:': 'i',
    'Ei': 'AI'
}

# Look through every line
for i in range(len(lines)):
    line = lines[i]
    first_char = line[0]
    second_char = line[line.find(' ')-1]

    # Check if the phoneme from Dutch transcription is in the en_phonemes
    if first_char not in en_phonemes:
        for key, value in replace_rules.items():
            if line.startswith(key):
                lines[i] = line.replace(key, value)
                break

# Save the file
with open('S3_task2_adjusted', 'w') as new_file:
    new_file.writelines(lines)
```
